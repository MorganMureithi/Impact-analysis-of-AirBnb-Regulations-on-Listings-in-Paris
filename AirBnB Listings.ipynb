```python
#Importing the dataset
Listings = pd.read_csv('C:\\Users\\hp\\Desktop\\Personal Projects\\Listings.csv',encoding='latin1',low_memory=False)
Listings
```

```python
#Reading the data
Listings
```

```python
#Accessing the column names
column_names = list(Listings.columns)
print(column_names)
```

```python
#Subsetting the 'host_since' date column
Host_since = Listings['host_since']
Host_since
#This is for the purpose of checking the datatype of host_since column
```

```python
#Changing the formart of the column type to datetime formart
Listings['host_since'] = pd.to_datetime(Listings['host_since'])
Listings['host_since']
```

```python
City_Listings = Listings['city'].unique()
City_Listings
#To identify the different cities in this dataset
```

```python
Paris_Listings = Listings[Listings['city']=='Paris'][['city','host_since','neighbourhood','accommodates','price']]
Paris_Listings
#To subset Paris listings only with time from when dwellers took up the space,their neighbourhood of choice,to the accommodation
#capacity of the space and price.
```

```python
#To check only Paris city was filtered
Only_Paris_Listings = Paris_Listings['city'].unique()
Only_Paris_Listings
```

```python
#To check for missing data
Missing_data = Paris_Listings.isnull().sum()
Missing_data
```

```python
# Droppping rows with missing values
New_Paris_Listings = Paris_Listings.dropna()
New_Paris_Listings
```

```python
New_Paris_Listings.describe()
#To obatin statistical summaries of numeric columns
```

```python
Paris_listings_neighbourhood = New_Paris_Listings.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
Paris_listings_neighbourhood
#To check pricing per neighbourhood,sorted from most expensive neighbourhood to the least expensive neighbourhood
```

```python
Paris_listings_accomodtions = New_Paris_Listings[New_Paris_Listings['neighbourhood'] == 'Elysee'].groupby('accommodates')['price'].mean().sort_values(ascending = False)
Paris_listings_accomodtions   
#To check the accomodation rates of Elysee estate-the most expensive neighbourhood
```

```python
Paris_listings_overtime = New_Paris_Listings.groupby(New_Paris_Listings['host_since'].dt.year).agg({'price':'mean','host_since': 'count'})
Paris_listings_overtime
# To check the hosting trend from when AirBnB began listig in Paris
```

```python
plt.barh(Paris_listings_neighbourhood.index,Paris_listings_neighbourhood.values,color = 'green',height = 0.5)
plt.title('Accomodation Prices by Neighbourhood')
plt.ylabel('neighbourhood')
plt.xlabel('price' + ('$'))
plt.show()
#A visual summary of the Accomodtaion pricing against neighbourhoods
```

```python
plt.barh(Paris_listings_accomodtions.index,Paris_listings_accomodtions.values, color = 'purple')
plt.title('Accomodation Rates in the  Most Expensive Neighbourhood')
plt.ylabel('Accomodatation' + ('PP'))
plt.xlabel('price' + ('$'))
plt.show()
#A visual presentation of of Accomodation capacity with rates
```

```python
plt.plot(Paris_listings_overtime['host_since'], linestyle = 'solid',color ='maroon',linewidth = '3')
plt.ylabel('Count of Hosts')
plt.xlabel('Year')
plt.title('No.of Hosts Across Time')
plt.show()
#A trendline of number of hosts overtime
```

```python
plt.plot(Paris_listings_overtime['price'], linestyle = 'solid', linewidth = '3',color = 'indigo' )
plt.ylabel('Average Price')
plt.xlabel('Year')
plt.title('Average Price across Years')
plt.show()
#A trend representation of Pricing across Years
```

```python
fig, ax1 = plt.subplots()
ax1.plot(Paris_listings_overtime['host_since'], linestyle = 'solid',color ='maroon',linewidth = '3')
ax1.set_ylabel('Count of Hosts',color = 'maroon')
ax1.set_xlabel('Year')
ax2 = ax1.twinx()
ax2.plot(Paris_listings_overtime['price'], linestyle = 'solid', linewidth = '3',color = 'indigo' )
ax2.set_ylabel('Average Price')
plt.title('Hosts Across Years w.rt to Price')
plt.show()
#An interaction of number of hosts overtime with Pricing
```
