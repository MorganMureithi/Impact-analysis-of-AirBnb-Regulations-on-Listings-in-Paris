```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Listings = pd.read_csv('LISTINGS.csv', encoding='latin1', low_memory=False)
Listings
```

```python
list(Listings.columns)
```

```python
print(Listings['host_since'].dtypes)
```

```python
Listings['host_since'] = pd.to_datetime(Listings['host_since'])
Listings['host_since']
```

```python
print(Listings['city'].unique())
```

```python
Paris_Listings = Listings[Listings['city'] == 'Paris'][['city','host_since','neighbourhood','accommodates','price']]
Paris_Listings
```

```python
print(Paris_Listings.isnull().sum())
```

```python
New_Paris_Listings = Paris_Listings.dropna()
New_Paris_Listings
```

```python
New_Paris_Listings.describe()
```

```python
Paris_listings_neighbourhood = New_Paris_Listings.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
Paris_listings_neighbourhood
```

```python
Paris_listings_accomodations = New_Paris_Listings[New_Paris_Listings['neighbourhood'] == 'Elysee'].groupby('accommodates')['price'].mean().sort_values(ascending=False)
Paris_listings_accomodations
```

```python
Paris_listings_overtime = New_Paris_Listings.groupby(New_Paris_Listings['host_since'].dt.year).agg({'price':'mean','host_since':'count'})
Paris_listings_overtime
```

```python
plt.barh(Paris_listings_neighbourhood.index,Paris_listings_neighbourhood.values,color='green')
plt.title('Accomodation Prices by Neighbourhood')
plt.ylabel('neighbourhood')
plt.xlabel('price' + '$')
plt.show()
```

```python
plt.barh(Paris_listings_neighbourhood.index,Paris_listings_neighbourhood.values,color='purple')
plt.title('Accomodation Rates in the Most Expensive Neighbourhood')
plt.ylabel('Accomodation')
plt.xlabel('price'+'$')
```

```python
plt.plot(Paris_listings_overtime['host_since'],linestyle='solid',color='maroon',linewidth=2)
plt.ylabel('Count of Hosts')
plt.xlabel('Year')
plt.title('No.of Hosts Across Time')
plt.show()
```

```python
plt.plot(Paris_listings_overtime['price'],linestyle='solid',color='purple',linewidth=2)
plt.ylabel('Average Price')
plt.xlabel('Year')
plt.title('Average Price across Years')
plt.show()
```

```python
fig,ax1 = plt.subplots()
ax1.plot(Paris_listings_overtime['host_since'],linestyle='solid',color='maroon',linewidth=2)
ax1.set_ylabel('Count of Hosts',color = 'maroon')
ax1.set_xlabel('Year')
ax2 = ax1.twinx()
ax2.plot(Paris_listings_overtime['price'],linestyle='solid',color='purple',linewidth=2)
ax2.set_ylabel('Average Price')
plt.title('Hosts Across year with respect to Price')
plt.show()
```
